```python
import pandas as pd
import numpy as np
from lifelines import CoxPHFitter
```

```python
# Load the CSV
df = pd.read_csv("../data/Telco_Customer_Churn_feature_engineered_v1.csv")

# Quick preview
df.head()
```

```python
survival_df = df[['tenure', 'Churn_num', 'MonthlyCharges', 'value_segment_Mid', 'value_segment_High', 'value_segment_VIP']]

cph = CoxPHFitter()
cph.fit(survival_df, duration_col='tenure', event_col='Churn_num')
cph.print_summary()
```
